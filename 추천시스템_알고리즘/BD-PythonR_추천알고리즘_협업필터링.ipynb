```python
def get_cosine_similarity(u, v):
    mask = np.isfinite(u) & np.isfinite(v)
    
    print( u, v )   # NaN ê±¸ëŸ¬ë‚´ê¸° ì´ì „
    
    u = u[mask]
    v = v[mask]
    
    print( u, v )   # NaN ê±¸ëŸ¬ë‚¸ ì´í›„
    
    uvdot = (u * v).sum()
    
    norm1 = (u ** 2).sum()
    norm2 = (v ** 2).sum()
    
    score = uvdot / np.sqrt(norm1 * norm2)
    
    return score
```

```python
from itertools import product

def get_cosine_similarity_table(ratings):
    index_combinations = list(product(ratings.index, repeat=2))
    
    similarity_list = []
    
    for uname, vname in index_combinations:
        u = ratings.loc[uname]
        v = ratings.loc[vname]
        
        score = get_cosine_similarity(u, v)
        
        similarity = {
            'u': uname,
            'v': vname,
            'score': score
        }
        
        similarity_list.append(similarity)
        
    similarity_list = pd.DataFrame(similarity_list)
    
    similarity_table = pd.pivot_table(similarity_list, index='u', columns='v', values='score')
    
    return similarity_table
```

```python
similarity_table = get_cosine_similarity_table(ratings)
```

```python
def predict_rating(user_name, book_name):
    neighbors_ratings = ratings[book_name].drop(index = user_name)  # ìì‹ ì˜ ê²ƒì€ ì œì™¸ì‹œí‚´
    neighbors_similarity = similarity_table[user_name].drop(index = user_name)
    
    nominator = (neighbors_similarity * neighbors_ratings).sum()
    denominator = neighbors_similarity.sum()
    
    score = nominator / denominator
    
    return score
```

### í‰ì  í…Œì´ë¸”

```python
print(len(list(product(ratings.index, ratings.columns))))
list(product(ratings.index, ratings.columns))
```

```python
def predict_rating_table(ratings):
    user_book_combinations = list(product(ratings.index, ratings.columns))
    
    rating_list = []
    
    for user_name, book_name in user_book_combinations:
        score = predict_rating(user_name, book_name)
        
        rating_predict = {
            'user': user_name,
            'book': book_name,
            'score': score
        }
        
        rating_list.append(rating_predict)
        
    rating_list = pd.DataFrame(rating_list)
    
    rating_table = pd.pivot_table(rating_list, index='user', columns='book', values='score')
    rating_table = rating_table[ratings.isnull()]
    
    return rating_table
```

```python
rating_table = predict_rating_table(ratings)
rating_table
```

### Pearson Correlation Coefficient

í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” Collaborative Filteringì—ì„œ ìœ ì‚¬ë„ ì¸¡ì • ë©”íŠ¸ë¦­ìœ¼ë¡œ ì£¼ë¡œ ì‚¬ìš©ë˜ë©°, ë‘ ë²¡í„°ì˜ ì„ í˜• ìƒê´€ê´€ê³„ë¥¼ í™•ì¸í•˜ê¸° ìœ„í•´ ì‚¬ìš©ë©ë‹ˆë‹¤. ì‹ìœ¼ë¡œ í‘œí˜„í•˜ìë©´ ì•„ë˜ì™€ ê°™ìŠµë‹ˆë‹¤.

$$
similarity(\textbf x,\textbf y)=r_{\textbf x\textbf y} = \frac{\sum (x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum (x_i - \bar x)^2}\sqrt{\sum (y_i - \bar y)^2}}
$$

í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” +1ê³¼ -1 ì‚¬ì´ì˜ ê°’ì„ ê°–ìŠµë‹ˆë‹¤. +1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ê°•í•œ ì–‘ì˜ ì„ í˜• ìƒê´€ê´€ê³„ê°€ ìˆë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•˜ê³  -1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ê°•í•œ ìŒì˜ ì„ í˜• ìƒê´€ê´€ê³„ê°€ ìˆë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ë˜í•œ 0ì€ ì•„ë¬´ëŸ° ì„ í˜• ìƒê´€ê´€ê³„ê°€ ì—†ìŒì„ ëœ»í•©ë‹ˆë‹¤. 

ìƒí’ˆ $i$ì™€ ìƒí’ˆ $j$ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” ì•„ë˜ì™€ ê°™ì´ ì •ì˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$
similarity(i,j)=\frac{\sum_{u \in U_{ij}} (r_{ui}-\bar{r_i}) \cdot (r_{uj}-\bar{r_j})}{\sqrt {\sum_{u \in U_{ij}} (r_{ui}-\bar{r_i})^2} \sqrt {\sum_{u \in U_{ij}} (r_{uj}-\bar{r_j})^2}}
$$

ë˜í•œ ì‚¬ìš©ì $u$ì™€ ì‚¬ìš©ì $v$ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” ì•„ë˜ì™€ ê°™ì´ ì •ì˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$
similarity(u,v)=\frac{\sum_{i \in I_{uv}} (r_{ui}- \bar {r_u}) \cdot (r_{vi}-\bar {r_v})}{\sqrt {\sum_{i \in I_{uv}} (r_{ui}- \bar {r_u})^2} \sqrt {\sum_{i \in I_{uv}} (r_{vi}-\bar {r_v})^2}}
$$

í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•  ë•Œ ì£¼ì˜í•´ì•¼ í•  ì ì´ ìˆìŠµë‹ˆë‹¤. ë°”ë¡œ ë‘ ë²¡í„°ê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ ê³„ì‚°í•˜ëŠ” ê³¼ì •ì—ì„œ ë¶„ëª¨ê°€ 0ì´ ë  ìˆ˜ ìˆë‹¤ëŠ” ì ì…ë‹ˆë‹¤. ë¶„ëª¨ê°€ 0ì´ë˜ëŠ” ê²½ìš° ë¶„ìë„ 0ì´ ë‚˜ì˜¤ëŠ”ë° ì´ëŸ¬í•œ ê²½ìš° ê²°ê³¼ê°€ 0ì´ ë˜ì–´ì•¼í•˜ì§€ë§Œ 0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ë‹¤ëŠ” ì—ëŸ¬ê°€ ë°œìƒí•©ë‹ˆë‹¤. ë”°ë¼ì„œ ë¶„ëª¨ê°€ 0ì´ë˜ì§€ ì•Šë„ë¡ ë§Œë“¤ë©° ë¶„ëª¨ê°€ 0ì´ ì•„ë‹ˆë”ë¼ë„ ê²°ê³¼ì— ì˜í–¥ì„ ì£¼ì§€ ì•ŠëŠ” ì ë‹¹íˆ ì‘ì€ ê°’ì„ ë”í•´ì£¼ì–´ì•¼ í•©ë‹ˆë‹¤. ì´ ë”í•˜ëŠ” ì‘ì€ ê°’ì„ ë³´í†µ **epsilon**ì´ë¼ê³  í•©ë‹ˆë‹¤. íŒŒì´ì¬ì—ì„œ **epsilon**ì€ ë³´í†µ 0.0001 ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

**get_pearson_correlation()**

```python
# ë‘ ë²¡í„°ê°„ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•´ì£¼ëŠ” get_pearson_correlation í•¨ìˆ˜ë¥¼ êµ¬í˜„í•´ì£¼ì„¸ìš”.
# ë‘ ë²¡í„° u, v ì¤‘ì— í•œ ê°œë¼ë„ ê°’ì´ ë¹„ì–´ìˆë‹¤ë©´ ê·¸ì— ëŒ€ì‘í•˜ëŠ” ê°’ë“¤ì€ ìœ ì‚¬ë„ë¥¼ êµ¬í•  ë•Œ ì œì™¸í•´ì£¼ì„¸ìš”.
# ë¶„ëª¨ê°€ 0ì´ ë˜ëŠ” ê²½ìš°ë¥¼ ì œì™¸í•˜ê¸° ìœ„í•´ ìµœì¢…ì ìœ¼ë¡œ ë¶„ëª¨ì— epsilonì„ ë”í•´ì£¼ì„¸ìš”.
def get_pearson_correlation(u, v, epsilon):
    # ë²¡í„° uì™€ vì¤‘ì— í•œ ê°œë¼ë„ ê°’ì´ ë¹„ì–´ìˆìœ¼ë©´
    # ëŒ€ì‘í•˜ëŠ” ìœ„ì¹˜ì— False, ë‘˜ ë‹¤ ê°’ì´ ì±„ì›Œì ¸ìˆë©´ Trueê°€ ë“¤ì–´ê°€ë„ë¡ maskë¥¼ ë§Œë“­ë‹ˆë‹¤.
    # np.isfinite()ëŠ” ì…ë ¥í•œ ë²¡í„°ì˜ ìœ„ì¹˜ë§ˆë‹¤ ê°’ì´ ë¹„ì–´ìˆìœ¼ë©´(NaN) False, ë¹„ì–´ìˆì§€ ì•Šìœ¼ë©´ Trueë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    mask = (np.isfinite(u) & np.isfinite(v))
    
    # ìœ„ì—ì„œ ë§Œë“  maskë¥¼ ì‚¬ìš©í•˜ì—¬ u, v ëª¨ë‘ ì±„ì›Œì ¸ìˆëŠ” ìœ„ì¹˜ì˜ ê°’ë“¤ë§Œ ì¸ë±ì‹±í•©ë‹ˆë‹¤.
    u, v = u[mask], v[mask]
    
    # u, v ê° ê° í‰ê· ê³¼ì˜ ì°¨ì´(í¸ì°¨)ë¥¼ êµ¬í•˜ì—¬ ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
    u_dev, v_dev = u-u.mean(), v-v.mean()
    # u, v ê° ê°ì˜ í¸ì°¨ì˜ ì œê³±ì˜ í•©ì„ êµ¬í•˜ì—¬ ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
    u_var, v_var = (u_dev**2).sum(), (v_dev**2).sum()
    
    # u_dev, v_dev ì˜ ë‚´ì (Dot Product)ì„ êµ¬í•˜ì—¬ ë¶„ìì— ì €ì¥í•©ë‹ˆë‹¤. ê° ì›ì†Œë¼ë¦¬ ê³±í•œ ê°’ë“¤ì˜ í•©ì„ êµ¬í•©ë‹ˆë‹¤.
    # u_dev.dot(v_dev) ë˜ëŠ” np.dot(u_dev, v_dev) ë¡œ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    nominator = (u_dev * v_dev).sum()
    # u_var, v_varì„ ê³±í•œë‹¤ëŠ” ë£¨íŠ¸ë¥¼ ì”Œì›Œ ë¶„ëª¨ë¥¼ êµ¬í•©ë‹ˆë‹¤.
    denominator = np.sqrt(u_var * v_var)
    
    # ë¶„ìì— ë¶„ëª¨ë¥¼ ë‚˜ëˆ„ì–´ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•©ë‹ˆë‹¤. ë‹¨ ë¶„ëª¨ê°€ 0ì¸ ê²½ìš°ë¥¼ ì²˜ë¦¬í•´ì£¼ê¸° ìœ„í•´ ë¶„ëª¨ì— epsilonì„ ë”í•©ë‹ˆë‹¤.
    pearson_correlation = nominator / (denominator + epsilon)
    
    return pearson_correlation    
```

```python
# ëª¨ë“  ì¡°í•©ì˜ ìˆ˜ë¥¼ êµ¬í•´ì£¼ëŠ” itertoolsì˜ productë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤. 
from itertools import product

# ëª¨ë“  ì‚¬ìš©ìê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•´ì£¼ëŠ” get_pearson_correlation_tableë¥¼ êµ¬í˜„í•´ì£¼ì„¸ìš”.
# ë¶„ëª¨ê°€ 0ì´ ë˜ëŠ” ê²½ìš°ë¥¼ ì œì™¸í•˜ê¸° ìœ„í•´ ìµœì¢…ì ìœ¼ë¡œ ë¶„ëª¨ì— epsilonì„ ë”í•´ì£¼ì„¸ìš”. epsilonì˜ ê¸°ë³¸ê°’ì€ 0.0001ìœ¼ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
def get_pearson_correlation_table(ratings, epsilon = 0.0001):
    # itertoolsì˜ productë¥¼ ì‚¬ìš©í•˜ì—¬ ì¸ë±ìŠ¤ì™€ ì¸ë±ìŠ¤ ìŒì˜ ëª¨ë“  ì¡°í•©ì„ êµ¬í•©ë‹ˆë‹¤. ì§€ê¸ˆì€ ì‚¬ìš©ìì˜ ì´ë¦„ì´ indexê°€ ë©ë‹ˆë‹¤.
    # productëŠ” ì…ë ¥í•˜ëŠ” repeat ì˜ ìˆ˜ì— í•´ë‹¹í•˜ëŠ” ìŒì„ ë§Œë“­ë‹ˆë‹¤. 3ì„ ì…ë ¥í•˜ë©´ 3ê°œì˜ ì¸ë±ìŠ¤ ì¡°í•©ì„ ìƒì„±í•©ë‹ˆë‹¤.
    # ê²°ê³¼ë¥¼ listë¡œ ë§Œë“¤ì–´ ì €ì¥í•©ë‹ˆë‹¤.
    index_combinations = list(product(ratings.index, repeat=2))

    # ì‚¬ìš©ìê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
    similarity_list = []

    for u_name, v_name in index_combinations:
        # ëª¨ë“  ì‚¬ìš©ìì— ëŒ€í•˜ì—¬ ê° ì‚¬ìš©ìì˜ í‰ì ì— í•´ë‹¹í•˜ëŠ” ë²¡í„°ë¥¼ u, v ì— ì €ì¥í•©ë‹ˆë‹¤.
        u = ratings.loc[u_name]
        v = ratings.loc[v_name]
    
        # ìœ„ì—ì„œ êµ¬í˜„í•œ get_pearson_correlation()ì„ ì‚¬ìš©í•˜ì—¬ u, v ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•©ë‹ˆë‹¤.
        score = get_pearson_correlation(u, v, epsilon)

        # ì‚¬ìš©ìì˜ ì´ë¦„ê³¼ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ ë‹´ëŠ” ë”•ì…”ë„ˆë¦¬ ë³€ìˆ˜ë¥¼ ë§Œë“­ë‹ˆë‹¤.
        similarity = {
            'u': u_name,
            'v': v_name,
            'score': score
        }
        
        # ìœ„ì—ì„œ ë§Œë“  ì‚¬ìš©ìì™€ ì‚¬ìš©ìê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ê°€ ë‹´ê¸´ ë”•ë…€ë„ˆë¦¬ ë³€ìˆ˜ë¥¼ similarity_listì— ì¶”ê°€í•©ë‹ˆë‹¤.
        similarity_list.append(similarity)  
    
    # ìœ„ì˜ forë¬¸ì´ ëë‚˜ë©´ similarity_listì—ëŠ” ëª¨ë“  ì‚¬ìš©ìê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ê°€ ë‚¨ê²¨ì ¸ ìˆìŠµë‹ˆë‹¤.
    # ì´ë¥¼ ê¹”ë”í•˜ê²Œ ì •ë¦¬í•˜ê¸° ìœ„í•˜ì—¬ íŒë‹¤ìŠ¤(Pandas)ì˜ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
    similarity_list = pd.DataFrame(similarity_list)
    
    # íŒë‹¤ìŠ¤ì˜ í”¼ë²— í…Œì´ë¸”ì„ í™œìš©í•˜ì—¬ ê²°ê³¼ë¥¼ ê¹”ë”í•˜ê²Œ ì •ë¦¬í•©ë‹ˆë‹¤.
    similarity_table= pd.pivot_table(similarity_list, index="u", columns="v", values="score")

    # similarity_tableë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    return similarity_table
```

### Get Nearest Neighbors

ìš°ì„  íŠ¹ì • ì‚¬ìš©ì ğ‘¢ì˜ ìƒí’ˆ ğ‘–ë¥¼ í‰ê°€í•  ë•Œ ì‚¬ìš©ì ğ‘¢ì˜ ì´ì›ƒì„ êµ¬í•˜ê² ìŠµë‹ˆë‹¤. ê¸°ë³¸ì ìœ¼ë¡œ ì‚¬ìš©ì ğ‘¢ì™€ ìœ ì‚¬ë„ê°€ ê°€ì¥ í° këª…ì´ ì´ì›ƒì´ ë©ë‹ˆë‹¤. í•˜ì§€ë§Œ ì¡°ê±´ì„ ì¢€ ë” ì¶”ê°€í•´ì•¼í•©ë‹ˆë‹¤. ê°€ë ¹ ì•„ì´í…œì— í‰ê°€ë¥¼ ë‚´ë¦¬ì§€ ì•Šê±°ë‚˜ ìœ ì‚¬ë„ê°€ ì ˆëŒ€ì ìœ¼ë¡œ ì‘ì€ ì‚¬ìš©ìëŠ” ì´ì›ƒì´ ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.

ë‹¤ìŒì˜ ì¡°ê±´ì— í•´ë‹¹í•˜ëŠ” ì´ì›ƒì„ êµ¬í•´ì£¼ì„¸ìš”.

- ì´ì›ƒì˜ ê¸°ì¤€ì€ ìœ ì‚¬ë„ê°€ ëœë‹¤.
- ìê¸° ìì‹ ì€ ì´ì›ƒì—ì„œ ì œì™¸í•œë‹¤.
- í•´ë‹¹ ì•„ì´í…œì— í‰ê°€ë¥¼ ë‚´ë¦° ì‚¬ëŒë“¤ì¤‘ì—ì„œ ìµœê·¼ì ‘ ì´ì›ƒì„ ê³ ë¥¸ë‹¤. ì¦‰, ì•„ì´í…œì— í‰ê°€ë¥¼ ë‚´ë¦°ì  ì—†ëŠ” ì‚¬ëŒë“¤ì€ ê³ ë ¤í•˜ì§€ ì•ŠëŠ”ë‹¤.
- ìœ ì‚¬ë„ê°€ 0 ì´ìƒì¸ ì‚¬ëŒë“¤ì¤‘ì—ì„œ ìµœê·¼ì ‘ ì´ì›ƒì„ ê³ ë¥¸ë‹¤. 
- ìœ ì‚¬ë„ê°€ 0ì´ìƒì¸ ì´ì›ƒì´ ì—†ë‹¤ë©´ ê·¸ë‚˜ë§ˆ ê°€ì¥ ìœ ì‚¬ë„ê°€ ê°€ì¥ í° í•œ ëª…ì„ ê³ ë¥¸ë‹¤.

ê²°ê³¼ì ìœ¼ë¡œ ì•„ì´í…œì„ ì‚° ì‚¬ëŒë“¤ì¤‘ì—ì„œë§Œ ì´ì›ƒì´ ì§€ì •ë˜ê¸° ë•Œë¬¸ì— **ì˜ˆì¸¡í•˜ë ¤ëŠ” ì•„ì´í…œë§ˆë‹¤ êµ¬í•´ì§€ëŠ” ì´ì›ƒì´ ë‹¬ë¼ì§ˆ ìˆ˜ ìˆìŠµë‹ˆë‹¤.**

**get_nearest_neighbors()**

```python
# íŠ¹ì • ì‚¬ìš©ì user_nameì˜ ìƒí’ˆ item_nameë¥¼ ì˜ˆì¸¡í•  ë•Œ user_nameì˜ këª…ì˜ ì´ì›ƒë“¤ì„ êµ¬í•˜ëŠ” get_nearest_neighborsì„ êµ¬í˜„í•©ë‹ˆë‹¤.
# ì´ì›ƒì„ êµ¬í•˜ë ¤ëŠ” ì‚¬ìš©ìì˜ ì´ë¦„(user_name), í•´ë‹¹ ìƒí’ˆì˜ ì´ë¦„(item_name), í‰ì  ë°ì´í„°(ratings), ì‚¬ìš©ì ë˜ëŠ” ìƒí’ˆì˜ ìœ ì‚¬ë„ í…Œì´ë¸”(similarity), êµ¬í•˜ë ¤ëŠ” ìµœëŒ€ ì´ì›ƒì˜ ìˆ˜(k)ë¥¼ ì¸ìë¡œ ë°›ìŠµë‹ˆë‹¤.
# ê²°ê³¼ëŠ” ì´ì›ƒë“¤ì˜ ëª©ë¡ì´ ë˜ì–´ì•¼í•©ë‹ˆë‹¤. ê²°ê³¼ë¥¼ ë¦¬ìŠ¤íŠ¸(List)ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.
# í‰ì ì´ ì¡´ì¬í•˜ëŠ” ì´ì›ƒì´ ì—†ë‹¤ë©´ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
def get_nearest_neighbors(user_name, item_name, ratings, similarity, k):
    # kë¥¼ ë„ˆë¬´ í¬ê²Œ ì…ë ¥í• ì‹œ ìµœëŒ€ê°’ì„ ìì‹ ì„ ì œì™¸í•œ ì´ ì¸ì›ìˆ˜ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
    # ìš°ì„  ìì‹ ì„ ì œì™¸í•œ ì´ ì¸ì›ìˆ˜ë¥¼ êµ¬í•©ë‹ˆë‹¤.
    max_neighbors_num = similarity.shape[0] - 1
    # kê°€ ì´ ì¸ì›ìˆ˜ë³´ë‹¤ í¬ë©´ kë¥¼ ì´ ì¸ì›ìˆ˜ë¡œ ëŒ€ì²´í•©ë‹ˆë‹¤.
    k = min(k, max_neighbors_num)
    
    # ìê¸° ìì‹  ì œì™¸í•œ ì‚¬ìš©ìë“¤ì˜ ìœ ì‚¬ë„ì™€ í‰ì  ë°ì´í„°ë¥¼ ë½‘ì•„ëƒ…ë‹ˆë‹¤.
    neighbors_similarity = similarity[user_name].drop(index=user_name)
    neighbors_ratings = ratings[item_name].drop(index=user_name)
    
    # í•´ë‹¹ ìƒí’ˆì— í‰ì ì„ ì¤€ ì ì´ ìˆëŠ” ì‚¬ìš©ìë“¤ì˜ ìœ ì‚¬ë„ì™€ í‰ì  ë°ì´í„°ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
    neighbors_similarity = neighbors_similarity[neighbors_ratings.notnull()]
    neighbors_ratings = neighbors_ratings[neighbors_ratings.notnull()]

    # í•´ë‹¹ ìƒí’ˆì— í‰ì ì„ ì¤€ ì‚¬ìš©ìê°€ ì—†ë‹¤ë©´ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    if neighbors_ratings.shape[0] == 0:
        # ë¹ˆ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
        return neighbors_ratings.index
        
    # ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ìœ„ì—ì„œ êµ¬í•´ì˜¨ ì‚¬ìš©ìì˜ ìœ ì‚¬ë„ neighbors_similarityë¥¼ ì •ë ¬í•©ë‹ˆë‹¤.
    neighbors_similarity = neighbors_similarity.sort_values(ascending=False)
    
    # ìœ ì‚¬ë„ê°€ 0 ì´ìƒì¸ ì‚¬ìš©ìë§Œ ë½‘ì•„ ì´ì›ƒ í›„ë³´ë¥¼ ë§Œë“­ë‹ˆë‹¤.
    candidates = neighbors_similarity[neighbors_similarity > 0]

    if (candidates.shape[0]) == 0:
        # ìœ ì‚¬ë„ê°€ 0ì´ìƒì¸ê²Œ ì—†ë‹¤ë©´, ì¦‰ candidates ì˜ í¬ê¸°ê°€ 0ì´ë¼ë©´ 
        # ê·¸ë‚˜ë§ˆ ìœ ì‚¬ë„ê°€ ê°€ì¥ í° ì‚¬ìš©ìë¥¼ ì´ì›ƒì´ ë©ë‹ˆë‹¤.
        nearest_neighbors = neighbors_similarity.head(1)
    else:
        # ìµœëŒ€ k ëª…ì˜ ì´ì›ƒì„ êµ¬í•©ë‹ˆë‹¤.
        nearest_neighbors = candidates.head(k)

    # êµ¬í•œ ì´ì›ƒì˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    return nearest_neighbors.index
```

### Average Correction

ìœ ì‚¬í•œ ìƒí’ˆ/ì‚¬ìš©ìì— ë” í° ê°€ì¤‘ì¹˜ë¥¼ ì£¼ì–´ íŠ¹ì • ìƒí’ˆì— ëŒ€í•œ ì‚¬ìš©ìì˜ ratings ê°’ì„ ì˜ˆì¸¡í•©ë‹ˆë‹¤. ìœ ì‚¬ë„ì— ë”°ë¼ ratingsì— ìœ ì‚¬ë„ê°€ ë†’ì€ kê°œì˜ ìƒí’ˆì— ë‚´ë ¤ì§„ ratingsì˜ í‰ê· /ìœ ì‚¬ë„ê°€ ë†’ì€ këª…ì˜ ì‚¬ìš©ìê°€ ë‚´ë¦° ratingsì˜ í‰ê· ì„ ëº€ ê°’ì˜ ê°€ì¤‘ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤. ì´ ê°’ì— ìƒí’ˆì— ë‚´ë¦° ratings ê°’/ì‚¬ìš©ìê°€ ë‚´ë¦° ratings ê°’ì˜ í‰ê· ì„ ìµœì¢…ì ìœ¼ë¡œ ë”í•˜ì—¬ ì˜ˆì¸¡í•©ë‹ˆë‹¤. ìƒí’ˆ/ì‚¬ìš©ìê°€ ë°›ê±°ë‚˜ í•˜ëŠ” ratings íŠ¹ì„±ì„ ë” ì˜ ë°˜ì˜í•©ë‹ˆë‹¤.


#### ì‚¬ìš©ì ê¸°ë°˜(ì‚¬ìš©ì-based) ê¸°ë²•
1. íŠ¹ì • ìœ ì‚¬ë„ ë©”íŠ¸ë¦­ì— ë”°ë¼ ì‚¬ìš©ì $u$ì™€ ìœ ì‚¬í•œ këª…(í˜¹ì€ ì´í•˜)ì˜ ì´ì›ƒì„ ê²°ì •í•©ë‹ˆë‹¤.
2. ìœ ì‚¬í•œ këª…ì˜ ì‚¬ìš©ìê°€ ìƒí’ˆ $i$ì— ë‚´ë¦° ratingsì— ê°ê° këª…ì˜ ì‚¬ìš©ìê°€ ë‚´ë¦° ratingsì˜ í‰ê· ì„ ëº€ ê°’ì— ëŒ€í•´ ìœ ì‚¬ë„ì— ë”°ë¥¸ ê°€ì¤‘ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤.
3. ì‚¬ìš©ì $u$ì´ ë‚´ë¦° ratings ê°’ì˜ í‰ê· ì„ ë”í•˜ì—¬ ê°’ì„ ë³´ì •í•´ì¤ë‹ˆë‹¤.

$$
\hat r_{ui} = \bar r_u + \frac{\sum_{v \in {N_i}^k(u)}{similarity(u,v)} \cdot (r_{vi}-\bar r_v)}{\sum_{v \in {N_i}^k(u)}{similarity(u,v)}}
$$


#### ìƒí’ˆ ê¸°ë°˜(ìƒí’ˆ-based) ê¸°ë²•
1. íŠ¹ì • ìœ ì‚¬ë„ ë©”íŠ¸ë¦­ì— ë”°ë¼ ìƒí’ˆ $i$ì™€ ìœ ì‚¬í•œ kê°œ(í˜¹ì€ ì´í•˜)ì˜ ìƒí’ˆì„ ê²°ì •í•©ë‹ˆë‹¤.
2. ìœ ì‚¬í•œ kê°œì˜ ìƒí’ˆì— ë‚´ë¦° ì‚¬ìš©ì $u$ì˜ ratingsì— ê°ê° kê°œì˜ ìƒí’ˆì— ë‚´ë ¤ì§„ ratingsì˜ í‰ê· ì„ ëº€ ê°’ì— ëŒ€í•´ ìœ ì‚¬ë„ì— ë”°ë¥¸ ê°€ì¤‘ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤.
3. ìƒí’ˆ $i$ì— ë‚´ë ¤ì§„ ratings ê°’ì˜ í‰ê· ì„ ë”í•˜ì—¬ ê°’ì„ ë³´ì •í•´ì¤ë‹ˆë‹¤.

$$
\hat r_{ui} = \bar r_i + \frac{\sum_{j \in {N_u}^k(i)}{similarity(i,j)} \cdot (r_{uj}-\bar r_j)}{\sum_{j \in {N_u}^k(i)}{similarity(i,j)}}
$$

**average_correction_predict_rating()**

```python
# íŠ¹ì • ì‚¬ìš©ìì˜ ìƒí’ˆì˜ í‰ì ì„ ì˜ˆì¸¡í•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” average_correction_predict_ratingë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.
# ì˜ˆì¸¡í•˜ë ¤ëŠ” ì‚¬ìš©ìì˜ ì´ë¦„(user_name)ê³¼ í•´ë‹¹ ìƒí’ˆì˜ ì´ë¦„(item_name), í‰ì  ë°ì´í„°(ratings), ì‚¬ìš©ì ë˜ëŠ” ìƒí’ˆì˜ ìœ ì‚¬ë„ í…Œì´ë¸”(similarity), êµ¬í•˜ë ¤ëŠ” ìµœëŒ€ ì´ì›ƒì˜ ìˆ˜(k), epsilonë¥¼ ì¸ìë¡œ ë°›ìŠµë‹ˆë‹¤.
# epsilonì˜ ê¸°ë³¸ê°’ì€ 0.0001ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
def average_correction_predict_rating(user_name, item_name, ratings, similarity, k, epsilon=0.0001):
    # ì¸ìë¡œ ë°›ì€ ì‚¬ìš©ìì˜ ì˜ˆì¸¡í•˜ë ¤ëŠ” ìƒí’ˆì— ëŒ€í•œ ì´ì›ƒì„ êµ¬í•©ë‹ˆë‹¤.
    nearest_neighbors = get_nearest_neighbors(user_name, item_name, ratings, similarity, k)
    
    # ê°€ê¹Œìš´ ì´ì›ƒì´ ì—†ë‹¤ë©´(íŠ¹ì • ì•„ì´í…œì— í‰ì ì´ í•œ ê°œë„ ì—†ë‹¤ë©´) ë¹ˆ ê°’(NaN)ì„ ë°˜í™˜í•©ë‹ˆë‹¤.
    if len(nearest_neighbors) == 0:
        return np.nan
    
    # ì‚¬ìš©ìì™€ ì‚¬ìš©ìì˜ ì´ì›ƒë“¤ì˜ ìœ ì‚¬ë„ë§Œ ì¶”ì¶œí•©ë‹ˆë‹¤.
    nn_similarity = similarity.loc[user_name, nearest_neighbors]
    # ì‚¬ìš©ìì™€ ì‚¬ìš©ìì˜ ì´ì›ƒë“¤ì˜ í‰ì  ë°ì´í„°ë§Œ ì¶”ì¶œí•©ë‹ˆë‹¤.
    nn_ratings = ratings.loc[nearest_neighbors, item_name]
    
    # ì´ì›ƒë“¤ì´ ì˜ˆì¸¡í•˜ë ¤ëŠ” ìƒí’ˆì— ë‚´ë¦° í‰ì ì˜ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤.
    neighbors_rating_mean = nn_ratings.mean()
    # ì˜ˆì¸¡í•˜ë ¤ëŠ” ì‚¬ìš©ìê°€ í•´ë‹¹ ìƒí’ˆì— ë‚´ë¦° í‰ì ì˜ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤.  
    u_rating_mean = ratings.loc[user_name].mean()
    
    # ì´ì›ƒë“¤ì´ ë‚´ë¦° í‰ì ì— ì´ì›ƒë“¤ì˜ í‰ì ì˜ í‰ê· ì„ ëº€ ê°’ê³¼ ì´ì›ƒë“¤ê³¼ì˜ ìœ ì‚¬ë„ì˜ ë‚´ì (Dot Product)ì„ êµ¬í•©ë‹ˆë‹¤.
    # íŒë‹¤ìŠ¤ ë°ì´í„°í”„ë ˆì„(DataFrame)ì˜ .dot() ì„ ì´ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    nominator = nn_similarity.dot(nn_ratings-neighbors_rating_mean)
    #print(nn_similarity)
    #print(nn_ratings)
    #print(neighbors_rating_mean)
    # ìœ ì‚¬ë„ì˜ í•©ì„ êµ¬í•©ë‹ˆë‹¤.
    denominator = nn_similarity.sum()
    
    # nominatorì— denominator ë‚˜ëˆ„ì–´ ê°€ì¤‘ í‰ê· ì„ êµ¬í•œ ë’¤ u_rating_meanì„ ë”í•˜ì—¬ ë³´ì •í•©ë‹ˆë‹¤.
    # ë‹¨ epsilonì„ ë”í•˜ì—¬ ë¶„ëª¨ê°€ 0ì´ë˜ëŠ” ê²½ìš°ë¥¼ ì²˜ë¦¬í•©ë‹ˆë‹¤.
    prediction = u_rating_mean + nominator / (denominator + epsilon)
    
    # ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    return prediction
```

```python
# ëª¨ë“  ë¹„ì–´ìˆëŠ” ê°’ì„ ì˜ˆì¸¡í•˜ëŠ” average_correction_predict êµ¬í˜„í•©ë‹ˆë‹¤.
# ë‹¨ ë¹„ì–´ìˆëŠ” ê°’ë§Œ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
# ì‚¬ìš©ì ê¸°ë°˜, ìƒí’ˆ ê¸°ë°˜ ê¸°ë²• ëª¨ë‘ ê°€ëŠ¥í•˜ë„ë¡ êµ¬í˜„í•©ë‹ˆë‹¤.
# kindì— ì‚¬ìš©ì ê¸°ë°˜ì¸ì§€ ìƒí’ˆê¸°ë°˜ì¸ì§€ ì—¬ë¶€ë¥¼ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ ì‚¬ìš©ì ê¸°ë°˜ìœ¼ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
# këŠ” 2, epsilonì€ 0.0001ì„ ê¸°ë³¸ê°’ìœ¼ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
def average_correction_predict(ratings, kind='user', k=2, epsilon=0.0001):
    # ìƒí’ˆ ê¸°ë°˜ ì˜ˆì¸¡ì´ë©´ í–‰ë ¬ì„ ë’¤ì§‘ì–´ì£¼ê³ (transpose) ë§ˆì§€ë§‰ì— ê²°ê³¼ë¥¼ ë‹¤ì‹œ ë’¤ì§‘ì–´ì„œ ë°˜í™˜í•©ë‹ˆë‹¤.
    if kind == 'item':
        ratings = ratings.T
    
    # ìœ ì‚¬ë„(í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜) í…Œì´ë¸”ì„ êµ¬í•©ë‹ˆë‹¤.
    similarity = get_pearson_correlation_table(ratings, epsilon)
    
    # ì˜ˆì¸¡í•œ ê²°ê³¼ë¥¼ ë‹´ëŠ” ë¦¬ìŠ¤íŠ¸ë¥¼ ë§Œë“­ë‹ˆë‹¤.
    prediction_list = []

    # ëª¨ë“  ì‚¬ìš©ìì˜ ëª¨ë“  ìƒí’ˆì— ëŒ€í•˜ì—¬ ì˜ˆì¸¡ì„ ì§„í–‰í•©ë‹ˆë‹¤.
    # itertoolì˜ productë¡œ ëª¨ë“  ì‚¬ìš©ìì™€ ëª¨ë“  ìƒí’ˆì˜ ì¡°í•©ì„ ë§Œë“­ë‹ˆë‹¤.
    for u_name, v_name in list(product(ratings.index, ratings.columns)):
        # ìœ„ì—ì„œ êµ¬í˜„í•œ average_correction_predict_ratingìœ¼ë¡œ ì‚¬ìš©ìì˜ ìƒí’ˆì„ ì˜ˆì¸¡í•©ë‹ˆë‹¤.
        score = average_correction_predict_rating(u_name, v_name, ratings, similarity, k, epsilon)
        
        # ì˜ˆì¸¡í•œ ê²°ê³¼ë¥¼ ë”•ì…”ë„ˆë¦¬ì— ì €ì¥í•©ë‹ˆë‹¤.
        prediction = {
            'u' : u_name,
            'v' : v_name,
            "score" : score
        }
        # prediction_list ì˜ˆì¸¡í•œ ê²°ê³¼ë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.
        prediction_list.append(prediction)

    # ê²°ê³¼ë¥¼ íŒë‹¤ìŠ¤ì˜ í”¼ë²— í…Œì´ë¸”ì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ê¹”ë”í•˜ê²Œ ì •ë¦¬í•©ë‹ˆë‹¤.
    result = pd.DataFrame(prediction_list).pivot_table(index='u', columns='v', values='score')
    # ë‹¨ ë¹„ì–´ìˆëŠ” ê°’ë§Œ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë‚¨ê¹ë‹ˆë‹¤.
    result = result[ratings.isnull()]
        
    # ì•„ì´í…œ ê¸°ë°˜ì´ë©´ ê²°ê³¼ë¥¼ ë‹¤ì‹œ ë’¤ì§‘ì–´ì¤ë‹ˆë‹¤.
    if kind == 'item':
        return result.T

    # ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    return result
```

## Recommend

- í‰ì ì˜ ê°œìˆ˜ê°€ ì§€ì •í•œ ê°œìˆ˜ ì´ìƒì¸ ìƒí’ˆë§Œ ì¶”ì²œí•œë‹¤.
- ì´ë¯¸ ë³¸ ê²ƒì€ ì¶”ì²œí•˜ì§€ ì•ŠëŠ”ë‹¤.

**recommend()**

```python
# recommendë¥¼ êµ¬í˜„í•´ì£¼ì„¸ìš”!
# numbers : ì¶”ì²œí•´ì¤„ ìµœëŒ€ ì±…ì˜ ê°œìˆ˜
# min_counts : ì¶”ì²œí•  ìƒí’ˆì´ ë°›ì€ í‰ì ì˜ ìµœì†Œ ê°œìˆ˜, ìµœì†Œ ì´ ê°œìˆ˜ ì´ìƒì˜ ì±…ë§Œ ì¶”ì²œí•©ë‹ˆë‹¤.
# numbers=3, min_count=3ë¥¼ ê¸°ë³¸ê°’ìœ¼ë¡œ ì§€ì •í•˜ê² ìŠµë‹ˆë‹¤.
def recommend(name, ratings, prediction, numbers=3, min_count=3):
    # ì…ë ¥í•œ ì‚¬ìš©ìì— ëŒ€í•œ ì˜ˆì¸¡ê°’ë§Œ ë‚¨ê¹ë‹ˆë‹¤. ì‹œë¦¬ì¦ˆ(Series) í˜•íƒœê°€ ë˜ë¯€ë¡œ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
    result = prediction.loc[name].to_frame()
    # ìƒí’ˆì´ ë°›ì€ í‰ì ì˜ ê°œìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” count ì»¬ëŸ¼ì„ ë§Œë“­ë‹ˆë‹¤.
    result['count'] = (ratings.notnull()).sum()
    # í‰ì ì´ ì íŒ ì»¬ëŸ¼ì˜ ì´ë¦„ì„ ratingìœ¼ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
    result.columns = ['rating', 'count']
    # í‰ì ì„ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•©ë‹ˆë‹¤.
    result = result.sort_values('rating', ascending=False)
    
    # ì´ë¯¸ ë³¸ ê²ƒì„ ì œì™¸í•˜ê³  ìµœì†Œ ì˜í™” í‰ê°€ ê°œìˆ˜ ì´ìƒì¸ ê²°ê³¼ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
    over_min_count = result['count'] >= min_count
    not_wathced = result['rating'].notnull()
    #print(over_min_count)
    #print(not_wathced)
    result = result[over_min_count&not_wathced]

    # ì¶”ì²œí•´ì¤„ ìµœëŒ€ ì±…ì˜ ê°œìˆ˜ë§Œí¼ ì¶”ì²œí•©ë‹ˆë‹¤.
    recommendation = result.head(numbers)

    # ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    return recommendation
```
